```python
df = pd.read_csv('Avatar_youtube.csv')
df.head()
```

Sort the data 2.b by top 10 comments in descending order and consider
the video IDs and Titles of top 10 videos which have highest comments

```python
df.rename(columns={'No, of Comments': 'No. of Comments'}, inplace=True)
```

```python
df.head()
```

```python
top_10_comments = df.sort_values(by='No. of Comments', ascending=False).head(10)
top_10_video_id  = top_10_comments['video ID'].tolist() # series -> list
top_10_video_titles = top_10_comments['Title'].tolist()
```

```python
from pprint import pprint
```

```python
comments_data = {}
for v_id ,v_title in zip(top_10_video_id, top_10_video_titles):
    comments_req = youtube.commentThreads().list(
        part = 'snippet',
        videoId = v_id
    )
    
    comments_response = comments_req.execute()

    # extracting comments
    comments = [item['snippet']['topLevelComment']['snippet']['textDisplay'] for item in comments_response['items']]
    comments_data[v_title] = comments

    pprint(comments_data)
```

```python
import json 
```

```python
with open('comments_output.json', 'w') as f:
    json.dump(comments_data, f)
    
print("Output exported to comments_output.json")



def likes_views_ratio(video_id):
    stats_req = youtube.videos().list(
        part = 'statistics',
        id = video_id
    )
    stats_rs = stats_req.execute()
    # extracting likes adn views count
    likes = int(stats_rs['items'][0]['statistics'].get('likeCount',0))
    views = int(stats_rs['items'][0]['statistics'].get('viewCount',0))
    ratio = likes/views
    return ratio


likes_view_r ={}
for video_id in top_10_video_id:
    ratio = likes_views_ratio(video_id)
    likes_view_r[video_id]= ratio
    
print("likes vs View Ratios:")
print(likes_view_r)
```
